```python
from __future__ import print_function

import pylab as pl
from scipy.optimize import curve_fit, minimize
    
%pylab inline
import numpy as np
import matplotlib.pyplot as pt
%pylab inline
```

Normal (Gaussian) Distribution

```python
#GENERATE 100 samples of different sizes N (N>10 & N<2000) from each of 5 different distributions (500 samples in total)
#distribution 1: gaussian
d1 = {}
# d1 is an empty disctionary
d1['gaussian'] = np.random.normal(loc=0.0, scale=1.0, size=100)
# d1 now is a dictionary with one object: an array by the key 'gaussian'
pl.hist(d1['gaussian'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

```python
#show mean and standard deviation for above distribution
print ("Gaussian mean: %.2f, standard deviation: %.2f"%(d1['gaussian'].mean(), d1['gaussian'].std()))
```

Repeat process for 100 distributions:

```python
#create array
d1_100 = np.linspace(5,2000,100)
print (d1_100, d1_100.shape)
```

```python
#redefining the dictionary
d1['gaussian'] = {} 

#create for loop 
for n in d1_100:
    d1['gaussian'][n] = np.random.normal(loc=0.0, scale=1.0, size=100)
#plot means of 100 distributions
d1['gaussian']['means'] = {}
d1_ax = pl.figure(figsize=(10,6)).add_subplot(111)

for x in d1['gaussian']:
    if not type(x) == str:
        d1['gaussian']['means'][x] = d1['gaussian'][x].mean()
        #and plot it
        d1_ax.plot(x, d1['gaussian']['means'][x], 'o')
        d1_ax.set_xlabel('sample size', fontsize=18)
        d1_ax.set_ylabel('sample mean', fontsize=18)
        d1_ax.set_title('Gaussian', fontsize=18)
        d1_ax.plot([min(d1_100), max(d1_100)],[0,0], 'k')
```

```python
# plot histogram of all means
d1_means = list(d1['gaussian']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(d1_means,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

Distribution 2: Chi Squared

```python
#distribution 2: chisq
d2mean = 100
df = d2mean

d2 = {}
# d2 is an empty dictionary
d2['chisq'] = np.random.chisquare(df, size=100)
# d2 now is a dictionary with one object: an array by the key 'chisq'

pl.hist(d2['chisq'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

```python
print ("Chisq mean: %.2f, standard deviation: %.2f"%(d2['chisq'].mean(), d2['chisq'].std()))
```

```python
#create array
d2_100 = np.linspace(5,2000,100)
print (d2_100, d2_100.shape)
```

```python
d2['chisq'] = {} 

for n in d2_100:
    d2['chisq'][n] = np.random.chisquare(df, size=100)

d2['chisq']['means'] = {}
d2_ax = pl.figure(figsize=(10,6)).add_subplot(111)

for x in d2['chisq']:
    if not type(x) == str:
        d2['chisq']['means'][x] = d2['chisq'][x].mean()
        
        d2_ax.plot(x, d2['chisq']['means'][x], 'o')
        d2_ax.set_xlabel('sample size', fontsize=18)
        d2_ax.set_ylabel('sample mean', fontsize=18)
        d2_ax.set_title('Chi Squared', fontsize=18)
        d2_ax.plot([min(d2_100), max(d2_100)],[df,df], 'k')
```

```python
# plot histogram of all means
d2_means = list(d2['chisq']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(d2_means,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

Distribution 4: Binomial

```python
#distribution 4: binomial
p = .5
n = 100
d4 = {}
# d4 is an empty disctionary
d4['bi'] = np.random.binomial(n, p, size=100)

# d4 now is a dictionary with one object: an array by the key 'pois'

pl.hist(d4['bi'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

```python
print ("Binomial mean: %.2f, standard deviation: %.2f"%(d4['bi'].mean(), d4['bi'].std()))
```

```python
#create array
d4_100 = np.linspace(5,2000,100)
print (d4_100, d4_100.shape)
```

```python
d4['bi'] = {} 
newp = .5
newn = 100
for n in d4_100:
    d4['bi'][n] = np.random.binomial(newn, newp, size=100)

d4['bi']['means'] = {}
d4_ax = pl.figure(figsize=(10,6)).add_subplot(111)

for x in d4['bi']:
    if not type(x) == str:
        d4['bi']['means'][x] = d4['bi'][x].mean()
        #and plot it
        d4_ax.plot(x, d4['bi']['means'][x], 'o')
        d4_ax.set_xlabel('sample size', fontsize=18)
        d4_ax.set_ylabel('sample mean', fontsize=18)
        d4_ax.set_title('Binomial', fontsize=18)
        d4_ax.plot([min(d4_100), max(d4_100)],[50,50], 'k')
```

```python
# plot histogram of all means
d4_means = list(d4['bi']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(d4_means,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```

Distribution 5: Standard Cauchy

```python
#distribution 5: Standard Cauchy

d5 = {}
# d5 is an empty disctionary
d5['sc'] = np.random.standard_cauchy(size=100)

# d5 now is a dictionary with one object: an array by the key 'pois'

pl.hist(d5['sc'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

```python
print ("Binomial mean: %.2f, standard deviation: %.2f"%(d5['sc'].mean(), d5['sc'].std()))
```

```python
#create array
d5_100 = np.linspace(5,2000,100)
print (d5_100, d5_100.shape)
```

```python
d5['sc'] = {} 

for n in d5_100:
    d5['sc'][n] = np.random.standard_cauchy(size=100)

d5['sc']['means'] = {}
d5_ax = pl.figure(figsize=(10,6)).add_subplot(111)

for x in d5['sc']:
    if not type(x) == str:
        d5['sc']['means'][x] = d5['sc'][x].mean()
        #and plot it
        d5_ax.plot(x, d5['sc']['means'][x], 'o')
        d5_ax.set_xlabel('sample size', fontsize=18)
        d5_ax.set_ylabel('sample mean', fontsize=18)
        d5_ax.set_title('Standard Cauchy', fontsize=18)
        d5_ax.plot([min(d5_100), max(d5_100)],[-0.2,-0.2], 'k')
```

```python
# plot histogram of all means
d5_means = list(d5['sc']['means'].values())

pl.figure(figsize=(10, 10))
pl.hist(d5_means,bins=30)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)
```
